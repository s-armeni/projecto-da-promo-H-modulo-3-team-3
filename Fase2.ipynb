4.1 Estandarización de la columna 'Age'.

````python
#Función para pasar los números de object a int, y los números formato 'forty-seven' a int
def age_to_int(num):
    try:
        return int(num)
    except ValueError:
        try:
            return w2n.word_to_num(num)
        except ValueError:
            return np.nan
df_data['Age']=df_data['Age'].apply(age_to_int)
````

4.4 Estandarización de la columna'Environment_satisfaction'.

````python
#Definimos una funcion:
def extrapolate(valor):
    if 4>=valor:
        return valor
    elif 4<valor<=20:
        return 1
    elif 20<valor<30:
        return 2
    elif 30<valor<40:
        return 3
    elif valor>40:
        return 4
    
#Aplicamos la funcion anterior a la columna 'Environment_satisfaction'
df_data['Environment_satisfaction']=df_data['Environment_satisfaction'].apply(extrapolate)
````

4.5 Eliminamos % de la columna 'Daily_rate'.

````python
#Función para eliminar los $
def remove_dollar(df, columns):
    for i in columns:
        df[i]=df[i].str.replace('$', '', regex=False)
    return df

#En esta lista están todas las columnas a las que tenemos que aplicar la función
list_dollar=['Daily_rate', 'Salary']

#Llamamos a la función para modificar la lista
df_data=remove_dollar(df_data, list_dollar)
````

4.6 Sustituimos 'Not Available' por 'NaN' en la columna 'Hourly_rate'

````python
#Para poder operar con la media y transformalos en float, es necesario pasar el 'not available' a valores nulo nan:
def change_to_null (string):
    if string == 'Not Available':
        return np.nan
    else:
        return string

df_data['Hourly_rate']=df_data['Hourly_rate'].apply(change_to_null)
````

4.8 Eliminamos posibles espacios en blanco al principio y final de las celdas.

````python
#Función para eliminar los espacios en blanco al principio y al final de los strings
def strip_strings(cell):
    if isinstance(cell, str): #si es un string retorna sin espacios
        return cell.strip()
    return cell

#Aplicamos la función a todas las celdas del DataFrame
df_data=df_data.applymap(strip_strings) #.applymap() porque lo aplico a cada elemento del df
````

4.9 Estandarizamos la columna de género.

````python
#Función para reemplazar 0 por M y 1 por F
def male_female(df, column):
    df[column]=df[column].replace({0:'Male', 1:'Female'})
    return df

#Aplicamos la función a la columna gender
df_data=male_female(df_data, 'Gender')
````

4.10 Estandarizamos las columnas de 'Yes/No'.

````python
#Función para unificar yes/no
def unify_yes_no(df, columns):
    yes_no_map={'yes':'Yes', 'YES':'Yes', 'TRUE':'Yes', 'true':'Yes', '1':'Yes', 'Y':'Yes', 'Yes':'Yes',
        'no':'No', 'No':'No', 'NO':'No', 'FALSE':'No', 'false':'No', '0':'No', 'False':'No'}
    
    for i in columns:
        df[i]=df[i].astype(str).map(yes_no_map)
    return df

#Llamamos a la función para modificar la lista
df_data=unify_yes_no(df=df_data, columns=list_yes_no)
````

4.11 Estandarizamos columnas integer a float.

````python
#Función para convertir int to float
def int_to_float(df, columns):
    for i in columns:
        df[i]=df[i].astype(float).round(2)
    return df
    
#Llamamos a la función para modificar la lista
df_data=int_to_float(df_data, list_int_float)
````

4.12 Estandarizamos columnas object a float.

````python
#Función para convertir object a float
def object_to_float(df, columns):
    for i in columns:
        try:
            df[i]=df[i].str.replace(',', '.').astype(float).round(2)
        except:
            df[i]=df[i].astype(float).round(2)
    return df

#Llamamos a la función para modificar la lista
df_data=object_to_float(df_data, list_object_float)
````

### 5. Gestión de nulos y eliminación de columnas prescindibles

5.1 Código para calcular nulos y sus %.

````python
#Separamos las columnas por categóricas y numéricas
cat_cols=df_data.select_dtypes(include=['O'])
num_cols=df_data.select_dtypes(include=['number'])

#Calculamos la cantidad y % de nulos de las columnas categóricas
cat_nulos_df=pd.DataFrame({'Column': cat_cols.columns, #trae los nombres de las columnas
                            'Nulos': cat_cols.isnull().sum(), #trae la suma de los nulos
                            '% Nulos': (cat_cols.isnull().mean() * 100).round(2)}) #trae el % de nulos
cat_nulos_df=cat_nulos_df[cat_nulos_df['Nulos'] > 0] #filtra para aquellos >0

#Calculamos la cantidad y % de nulos de las columnas numéricas
num_nulos_df=pd.DataFrame({'Column': num_cols.columns, #trae los nombres de las columnas
                            'Nulos': num_cols.isnull().sum(), #trae la suma de los nulos
                            '% Nulos': (num_cols.isnull().mean() * 100).round(2)}) #trae el % de nulos

#Creamos variables en las que guardamos las medias y medianas de cada columna numérica
num_mean=num_cols.mean()
num_median=num_cols.median()
#Añadimos las medias y medianas al df de columnas numéricas
num_nulos_df['Mean']=num_nulos_df['Column'].map(num_mean)
num_nulos_df['Median']=num_nulos_df['Column'].map(num_median)

num_nulos_df=num_nulos_df[num_nulos_df['Nulos'] > 0] #filtra para aquellos >0
````

Columna Employee_number:

Tiene un 26.70% nulos. Para que la empresa pueda seguir utilizando este identificador a nivel interno, decidimos mantener esta columna pero, puesto que utilizaremos un ID autoincremental para nuestro análisis, convertimos la columna a object y sustituimos los NaN con 'unknown'.

````python
#Convertimos la columna a object
df_data['Employee_number']=df_data['Employee_number'].astype('object')

#Sustituimos los nullos por 'unkown'.
df_data['Employee_number']=df_data['Employee_number'].fillna('Unknown')
````

Para las columnas con un % de nulos inferior al 10%, aplicaremos la sustitución de esos nulos por la media o la mediana. Comparando éstas, y al ser tan parecidas entre sí, optamos por sustituir en base a la media.

````python
columns_mean=['Daily_rate', 'Hourly_rate', 'Work_life_balance']
for i in columns_mean:
    imputer=SimpleImputer(strategy='mean')
    df_data[i]=imputer.fit_transform(df_data[[i]])
````

Para las columnas con un % de nulos superior al 10% aplicamos un IterativeImputer (random_state=42) para sustituirlos con una estimación lo más precisa posible.

````python
#Reemplazamos nulos con IterativeImputer
imputer=IterativeImputer(max_iter=20, random_state=42)
imputed=imputer.fit_transform(df_data[['Performance_rating', 'Total_working_years']])

df_data[['Performance_rating', 'Total_working_years']]=imputed
````
