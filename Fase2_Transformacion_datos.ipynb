- 4.x Columna over_18. Es un object con un único valor Y (o vacío). Sustituimos los Y por True y lo pasamos a  boolean.

```python
# 🗑️
df_col_3['over_18'].unique()
```

```python
#Columna over_18. Es un object. La pasamos a boolean
#1. Convertimos los Y en True
df_col_3['over_18']=df_col_3['over_18'].replace('Y', True)

#2. Convertimos a boolean, rellenando vacíos con NaN
df_col_3['over_18']=df_col_3['over_18'].replace('', np.nan).astype('boolean')
```

```python
# 🗑️
#Comprobamos
df_col_3['over_18'].head(3)
```

- 4.x Columna over_time. Es un object con valores No, Yes y vacío. Sustituimos por True/False y lo pasamos a boolean.

```python
# 🗑️
df_col_3['over_time'].unique()
```

```python
#Columna over_time. Es un object. La pasamos a boolean
df_col_3['over_time']=df_col_3['over_time'].replace('No', False)
df_col_3['over_time']=df_col_3['over_time'].replace('Yes', True)
df_col_3['over_time']=df_col_3['over_time'].replace('', np.nan).astype('boolean')
```

```python
# 🗑️
#Comprobamos
df_col_3['over_time'].unique()
```

- 4.x Columna percent_salary_hike. Es un integer. Lo convertimos a float de 2 decimales.

```python
# 🗑️
df_col_3['percent_salary_hike'].dtype
```

```python
#Columna percent_salary_hike. Es un integer. Lo pasamos a float 2 decimales
#1. Lo convertimos en float
df_col_3['percent_salary_hike']=df_col_3['percent_salary_hike'].astype(float)

#2. Formateamos para su visualización a 2 decimales
df_col_3['percent_salary_hike']=df_col_3['percent_salary_hike'].map('{:.2f}'.format).astype(float)
```

```python
# 🗑️
#Comprobamos
df_col_3['percent_salary_hike'].dtype
```

- 4.x Columna performance_rating. Es un object. La convertimos en float.

- 4.x Columna standard_hours. Es un object. Convertimos a float 2 decimales

```python
# 🗑️
df_col_3['standard_hours'].unique()
```

```python
#Columna standard_hours. Es un object. Lo pasamos a float
#1. Pasamos las , a .
df_col_3['standard_hours']=df_col_3['standard_hours'].str.replace(',', '.')

#2. Convertimos a float
df_col_3['standard_hours']=df_col_3['standard_hours'].astype(float)

#3. Formateamos para su visualización a 2 decimales
df_col_3['standard_hours']=df_col_3['standard_hours'].map('{:.2f}'.format).astype(float)
```

- 4.x Columna total_working_years. Es un object. Convertimos a integer.

```python
# 🗑️
df_col_3['total_working_years'].unique()
```

```python
#Columna total_working_years. Es un object. Lo pasamos a integer (años completos)
#1. Eliminamos las ,
df_col_3['total_working_years']=df_col_3['total_working_years'].str.split(',').str[0]

#2. Convertimos a integer: pd.to_numeric convierte los nan en números y errors='coerce' coge esos números (previos nan) y los convierte en NaN.
df_col_3['total_working_years']=pd.to_numeric(df_col_3['total_working_years'], errors='coerce').astype('Int64')
```
